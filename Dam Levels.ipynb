```python
import csv
import pandas as pd
import json

filename = 'Dam levels update 2012-2017.csv'
df = pd.read_csv(filename, encoding = "ISO-8859-1")
print(df)
```

```python
def parse_chunk(df, start_index):
    cols = 4
    print(start_index)
    data = df.iloc[:, start_index:start_index + cols]
    dam_name = data.iloc[1][0]
    storage_data = data.iloc[:, 1][4:]
    storage_data = storage_data.str.replace('\s+', '')  # strip spaces
    storage_data = pd.to_numeric(storage_data, errors='coerce')
    return {
        'label': dam_name,
        'data': storage_data.tolist()
    }

print(parse_chunk(df, 10 * 4 + 1))
```

```python
all_data = [parse_chunk(df, 4 * i + 1) for i in range(13)]
```

```python
dates = df.iloc[:, 0][4:]
dates = pd.to_datetime(dates)
# dates = dates.apply(lambda x: x.to_
dates = [str(d) for d in dates.tolist()]
```

```python
data = {
    'labels': dates,
    'datasets': all_data
}
print(json.dumps(data, indent=2))
```
